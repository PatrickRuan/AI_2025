```python
import openai

# 設定金鑰與伺服器
openai.api_key = "ollama"
model = "gemma3:4b"
from openai import OpenAI
client1 = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
client2 = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

# 定義系統訊息
sys_msg_support = "你是一個支持廢除死刑的人，請基於這個立場回答問題。"
sys_msg_against = "你是一個反對廢除死刑的人，請基於這個立場回答問題。"

# 初始化訊息
messages1 = [{"role": "system", "content": sys_msg_support}]
messages2 = [{"role": "system", "content": sys_msg_against}]

# 模擬對話函數
def debate_round(client1, client2, messages1, messages2):
    # 模型1發表觀點
    response1 = client1.chat.completions.create(
        model=model,
        messages=messages1,
        temperature=0.7)
    reply1 = response1.choices[0].message.content
    print("支持廢死方:", reply1)

    # 將模型1的回應加入模型2的對話
    messages2.append({"role": "user", "content": reply1})

    # 模型2發表反對觀點
    response2 = client2.chat.completions.create(
        model=model,
        messages=messages2,
        temperature=0.7
    )
    reply2 = response2.choices[0].message.content
    print("反對廢死方:", reply2)

    # 將模型2的回應加入模型1的對話
    messages1.append({"role": "user", "content": reply2})

# 模擬多輪對話
for i in range(5):  # 模擬5輪對話
    print(f"\n第 {i + 1} 輪對話")
    debate_round(client1, client2, messages1, messages2)
```

```python
import openai

# 設定金鑰與伺服器
openai.api_key = "ollama"
from openai import OpenAI

# 建立 client1, client2, 和 gemma2
client1 = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")  # llama3.2
client2 = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")  # mistral:7b
client_judge = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")  # gemma2

# 定義系統訊息
sys_msg_support = "你是一個支持廢除死刑的人，請基於這個立場回答問題。而且總是使用繁體中文。"
sys_msg_against = "你是一個反對廢除死刑的人，請基於這個立場回答問題。而且總是使用繁體中文。"

# 初始化訊息
messages1 = [{"role": "system", "content": sys_msg_support}]
messages2 = [{"role": "system", "content": sys_msg_against}]

# 模擬對話函數
def debate_round(client1, client2, messages1, messages2):
    # llama3.2 發表觀點
    response1 = client1.chat.completions.create(
        model="llama3.2:latest",
        messages=messages1,
        temperature=0.7
    )
    reply1 = response1.choices[0].message.content
    print("支持廢死方 (llama3.2):", reply1)

    # 將模型1的回應加入模型2的對話
    messages2.append({"role": "user", "content": reply1})

    # mistral:7b 發表反對觀點
    response2 = client2.chat.completions.create(
        model="mistral:7b",
        messages=messages2,
        temperature=0.7
    )
    reply2 = response2.choices[0].message.content
    print("反對廢死方 (mistral:7b):", reply2)

    # 將模型2的回應加入模型1的對話
    messages1.append({"role": "user", "content": reply2})

# 模擬多輪對話
for i in range(5):  # 模擬5輪對話
    print(f"\n第 {i + 1} 輪對話")
    debate_round(client1, client2, messages1, messages2)

# 評審階段
def judge_debate(client_judge, messages1, messages2):
    # 將雙方的完整對話歷史合併，並請 gemma2 判定勝負
    debate_summary = "以下是支持廢死和反對廢死雙方的完整辯論歷史，請作為評審判定誰在辯論中表現更好，並給予理由。而且總是使用繁體中文。"
    debate_messages = messages1 + messages2
    debate_messages.append({"role": "system", "content": debate_summary})

    response_judge = client_judge.chat.completions.create(
        model="gemma2",
        messages=debate_messages,
        temperature=0.5
    )
    judge_reply = response_judge.choices[0].message.content
    print("\n評審 (gemma2):", judge_reply)

# 呼叫評審
judge_debate(client_judge, messages1, messages2)
```
