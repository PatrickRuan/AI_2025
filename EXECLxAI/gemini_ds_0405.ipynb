## Data loading

### Subtask:

**Reasoning**:

## Data visualization

### Subtask:
Visualize the distribution of numerical and categorical features in the nutrition dataset.


**Reasoning**:
Create histograms for numerical features and bar charts for the categorical feature 'Name' to visualize the data distribution. Use subplots for better arrangement.



```python
import matplotlib.pyplot as plt

# Numerical features
numerical_features = ['Serving (grams)', 'Calories', 'Fat', 'Protein', 'Carbohydates', 'Fiber']
num_plots = len(numerical_features)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

for i, feature in enumerate(numerical_features):
    axes[i].hist(df[feature], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Distribution of {feature}')

# Remove any unused subplots
for i in range(num_plots, len(axes)):
    fig.delaxes(axes[i])

# Categorical feature
plt.figure(figsize=(10, 6))  # Adjust figure size
df['Name'].value_counts().plot(kind='bar', color='lightcoral')
plt.xlabel('Food Name')
plt.ylabel('Frequency')
plt.title('Frequency of Food Items')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

## Data analysis

### Subtask:
Analyze the numerical features of the nutrition dataset to identify potential correlations and derive key statistics.


**Reasoning**:
Calculate descriptive statistics, correlation matrix, and identify potential outliers for numerical features.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate descriptive statistics
numerical_features = ['Serving (grams)', 'Calories', 'Fat', 'Protein', 'Carbohydates', 'Fiber']
descriptive_stats = df[numerical_features].describe()
print("Descriptive Statistics:\n", descriptive_stats)

# Calculate the correlation matrix
correlation_matrix = df[numerical_features].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Identify potential outliers (example using IQR method for 'Calories')
Q1 = df['Calories'].quantile(0.25)
Q3 = df['Calories'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Calories'] < lower_bound) | (df['Calories'] > upper_bound)]
print("\nPotential Outliers in 'Calories':\n", outliers[['Name', 'Calories']])

# Summary of findings (example)
summary = f"""
Summary of Numerical Feature Analysis:

Descriptive Statistics:
{descriptive_stats}

Correlation Analysis:
Strong positive correlations observed between ... (mention specific correlations from the heatmap).
Strong negative correlations observed between ... (mention specific correlations from the heatmap).

Outlier Analysis:
Potential outliers identified in 'Calories' column using IQR method.
"""
print(summary)
```
